````python
%config InlineBackend.figure_formats = ['svg']
import xyzpy as xyz
import combo_dynamics
from math import pi

import os

# set single thread
os.environ['OMP_NUM_THREADS'] = '1'
os.environ['MKL_NUM_THREADS'] = '1'
os.environ['OPENBLAS_NUM_THREADS'] = '1'
# specify which gpu
os.environ['CUDA_VISIBLE_DEVICES'] = '3'
````

relevant measurements:

```python

measure = 'X37,41,52,56,57,58,62,79,Y75,Z38,40,42,63,72,80,90,91'
```

example backends:

```python
backend='numpy-double'  # numpy double precision
backend='torch-cpu-single'  # torch cpu single precision
backend='torch-gpu-single'  # torch cuda single precision
backend='torch-cpu-viagpu-double'  # torch cuda double precision, offloading storage to cpu
```

example optimizers:
```python
optimize = 'auto'  # cheap
optimize = 'auto-hq'  # more effort

# use predefined optimizer with dynamic slicing to set size, paths cached to disk
# see also combo_dynamics.get_optimizer_exact
optimize = None, target_size=2**27  
```

# MIX Method

````python
combo_dynamics.run_bpmixed(
    steps=9,
    theta_h=0.7,
    measure='Z62',
    chi=64,
    backend='torch-gpu-single',
    optimize='auto',
    progbar=True,
)
````

# PEPO Method

````python
combo_dynamics.run_bppepo(
    steps=9,
    theta_h=0.7,
    measure='Z62',
    chi=64,
    backend='torch-gpu-double',  # PEPO method a bit unstable (due to cutoff?) with single precision
    optimize='auto',
    progbar=True,
)
````

# PEPS Method

````python
combo_dynamics.run_bppeps(
    steps=9,
    theta_h=0.7,
    measure='Z62',
    chi=64,
    backend='torch-gpu-single',
    optimize='auto',
    progbar=True,
)
````

# Exact

````python
combo_dynamics.run_exact(
    steps=5, 
    theta_h=0.7,
    measure='X37,41,52,56,57,58,62,79,Y75,Z38,40,42,63,72,80,90,91',
    backend='torch-double-gpu',
    progbar=True,
)
````

# Using gpu with cpu offloading & dynamic slicing (for very large $\chi$)

````python
combo_dynamics.run_bpmixed(
    # n.b. at this number of steps not necessary (chi won't reach large)
    steps=9,
    theta_h=0.7,
    measure='Z62',
    chi=256,
    backend='torch-cpu-viagpu-single',
    # with the built in dynamic slicing optimizer, quite a lot of time 
    #   will be spent finding contraction paths (and not contracting...)
    optimize=None,  
    target_size=2**27,
    progbar=True,
)
````

# quick check relevant quantities:

````python
runner = xyz.Runner(combo_dynamics.run_bpmixed, None)
ds = runner.run_combos(
    cases=[
        {'steps': 5, 'measure': 'X13,29,31,Y9,30,Z8,12,17,28,32'},
        {'steps': 5, 'measure': 'X37,41,52,56,57,58,62,79,Y38,40,42,63,72,80,90,91,Z75'},
        {'steps': 5, 'measure': 'X37,41,52,56,57,58,62,79,Y75,Z38,40,42,63,72,80,90,91'},
        {'steps': 20, 'measure': 'Z62'},
    ],
    combos={
        'theta_h': [i * pi / 32 for i in range(0, 17)],
        'chi': [8, 16],
    },
    constants={
        'optimize': 'auto',  # use cheap optimizer
        'backend': 'torch-cpu-single',
        'progbar': False,
    },
    num_workers=64,
)
````

````python
ds.xyz.infiniplot(
    x='theta_h',
    y='O',
    color='chi',
    marker='chi',
    row=('steps', 'measure'),
)
````

````python
# times for a single cpu core
ds.xyz.infiniplot(
    x='theta_h',
    y='time', yscale='log',
    hue='chi',
    row=('steps', 'measure'),
)
````
